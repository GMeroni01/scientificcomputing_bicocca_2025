The function enumerate(sequence) returns tuples containing indices of objects in the sequence, and the objects.
The random module provides tools for working with the random numbers. In particular, random.randint(start, end) generates a random number not smaller than start, and not bigger than end.
- Generate a list of 10 random numbers from 0 to 9.
- Using the enumerate(random_list) function, iterate over the tuples of random numbers and their indices, and print out "Match: NUMBER and INDEX" if the random number and its index in the list match.



```python
# Part 1
names = ["Mary", "John", "Sarah"]
age = [21, 56, 98]

for n,a in zip(names,age):
    print(f"{n} is {a} years old")

# Part 2
import random

random_list = []

for _ in range(10):
    random_list.append(random.randint(0, 9))
    
for idx, num in enumerate(random_list):
    if idx == num:
        print(f"Match: {num} and {idx}")
```

## Exercise 12
- Write a function to return the first n rows of Pascal's triangle. The return should be 
a list of length n, with each element itself a list containing the numbers for that row.
- Write a function to print out Pascal's triangle with proper formatting, so the numbers 
in each row are centered between the ones in the row above.

```python
# Create Pascal's triangle
def pascal_triangle(n):
    triangle = []

    for i in range(n):
        row = []

        for j in range(i+1):
            if j == 0 or j == i:
                row.append(1)
            else:
                value = triangle[i-1][j-1] + triangle[i-1][j]
                row.append(value)

        triangle.append(row)

    return triangle

# Build a string from a list
def create_row(row):
    string = ""
    
    for i in range(len(row)):
        string += str(row[i])
        if i < len(row) - 1:
            string += "   "
            
    return string

# Print Pascal's triangle
def print_triangle(n):
    triangle = pascal_triangle(n)

    last_row = create_row(triangle[-1])

    for row in triangle:
        row_str = create_row(row)

        space = len(last_row) - len(row_str)
        left_space = space // 2

        for _ in range(left_space):
            print(" ", end="")

        print(row_str)
        
print_triangle(10)
```


## Exercise 13
We want to keep a schedule of events. We will do this by creating a class called Day. A Day holds a list of events and has methods that allow you to add an delete events. 
Our events will be instances of a class Event, which holds the time, location, and description of the event.
Finally, we can keep track of a list of all the Days for which we have events to make our schedule.
Fill in these classes and write some code to demonstrate their use:
- Create a full week of days in your calendar
- Add an event every day at noon called "lunch"
- Randomly add some other events to fill out your calendar
- Write some code that tells you the start time of your first meeting and the end time of your last meeting (this is the length of your work day)


```python
import random

class Event:
    """a single event in our calendar"""

    def __init__(self, name, time=9, location=None, duration=1):
        self.name = name
        self.time = time
        self.location = location
        self.duration = duration

    def end_time(self):
        return self.time + self.duration

class Day:
    """a single day keeping track of the events scheduled"""

    def __init__(self, month, day, year):
        self.month = month
        self.day = day
        self.year = year
        self.events = []

    def add_event(self, name, time=9, location=None, duration=1):
        event = Event(name, time, location, duration)
        self.events.append(event)

    def delete_event(self, name):
        for e in self.events:
            if e.name == name:
                self.events.remove(e)
                break

    def print_day(self):
        print(f"\nDay {self.day}/{self.month}/{self.year}")
        for e in sorted(self.events, key=lambda x: x.time):
            print(f"{e.time}:00 - {e.name} ({e.duration}h)")

    def work_day_duration(self):
        start = min(e.time for e in self.events)
        end = max(e.end_time() for e in self.events)
        return end - start

# Calendar filling

days = []
event_names = ["measurements", "meeting", "class", "study", "workout"]

for i in range(7):
    days.append(Day(4, 20 + i, 2026))

for day in days:
    # Morning event
    name = random.choice(event_names)
    time = random.randint(8, 11)
    duration = random.randint(1, 12-time)
    day.add_event(name, time, duration=duration)

    # Lunch
    day.add_event("lunch", time=12, duration=1)

    # Afternoon event
    name = random.choice(event_names)
    time = random.randint(13, 18)
    duration = random.randint(1, 3)
    day.add_event(name, time, duration=duration)

    day.print_day()
    print("Work duration:", day.work_day_duration(), "hours")
```
