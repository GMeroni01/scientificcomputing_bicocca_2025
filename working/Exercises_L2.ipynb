## Exercise 2
Here we will read in columns of numbers from a file and create a histogram, using NumPy routines. Make sure you have the data file "sample.txt" in the same directory as this notebook. You download it from https://raw.githubusercontent.com/sbu-python-summer/python-tutorial/master/day-3/sample.txt.
- Use python to download a file! Check out the requests package.
- Use np.loadtxt() to read this file in.
- Next, use np.histogram() to create a histogram array. The output returns both the count and an array of edges.
- Finally, loop over the bins and print out the bin center (averaging the left and right edges of the bin) and the count for that bin.

```python
import numpy as np
import requests

# Download file
url = "https://raw.githubusercontent.com/sbu-python-summer/python-tutorial/master/day-3/sample.txt"

with open("sample.txt", "w") as f:
    f.write((requests.get(url)).text)

# Histogram
data = np.loadtxt("sample.txt")

bins = 10
hist = np.histogram(data[:,1], bins)

for i in range(bins):
    center = (hist[1][i+1] + hist[1][i])/2
    print(f"Bin center: {center:.2f} \t Counts: {hist[0][i]}")
```

## Exercise 3
Write a function to calculate the standard deviation for an input array a:
- First compute the average of the elements in a to define ā
- Next compute the sum over the squares of a - ā
- Then divide the sum by the number of elements in the array
- Finally take the square root (you can use np.sqrt())
Test your function on a random array, and compare to the built-in np.std(). Check the runtime as well.

```python
import numpy as np
import time

def std_user(a):
    
    mean = np.sum(a)/len(a)
    sum_square_diff = np.sum((a - mean)**2)
    variance = sum_square_diff/len(a)

    return np.sqrt(variance)

a = np.random.random(100000)

# User
start = time.perf_counter()
std_user_value = std_user(a)
end = time.perf_counter()
user_time = end - start

# Numpy
start = time.perf_counter()
std_numpy = np.std(a)
end = time.perf_counter()
numpy_time = end - start

print(f"Std (user) = {std_user_value}")
print(f"Std (np.std) = {std_numpy}")
print(f"Difference of std between the two methods: {abs(std_user_value-std_numpy)}\n")

print(f"Runtime std_user function: {user_time:.5f} s")
print(f"Runtime np.std: {numpy_time:.5f} s")
print(f"--> np.std is {user_time/numpy_time :.2f} times faster than std_user function !")
```
