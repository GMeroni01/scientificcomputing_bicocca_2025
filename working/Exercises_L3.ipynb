## Exercise 3
- Plot y vs x.
- Draw a circle filled in with a solid color.
Generalize your circle drawing commands to produce a function "draw_circle(x0, y0, R, color)" that draws the circle. 
Here, (x0, y0) is the center of the circle, R is the radius, and color is the color of the circle.
- Now randomly draw 10 circles at different locations, with random radii, and random colors on the same plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def draw_circle(x0, y0, R, color):
    
    theta = np.linspace(0, 2*np.pi, 300)

    x = x0 + R*np.cos(theta)
    y = y0 + R*np.sin(theta)

    plt.plot(x, y, color=color)
    plt.fill(x, y, color=color, alpha=0.5)

plt.figure(figsize=(5,5))

for _ in range(10):
    
    x0 = random.uniform(-10, 10)   
    y0 = random.uniform(-10, 10)   
    R = random.uniform(0.5, 4)     

    color = (random.random(),random.random(),random.random()) # RGB combination
    
    draw_circle(x0, y0, R, color)

plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.title("Pois pattern")
plt.grid()
plt.show()
```

## Exercise 5
Create an x array using NumPy with a number of points, spanning from 0 to 2pi.
Create 3 axes vertically, and do the following:
- Define a new numpy array f initialized to a function of your choice
- Plot f in the top axes
- Compute a numerical derivative of f and plot this in the middle axes
- Compute the second derivative and plot that in the bottom axes

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2*np.pi, 500)
a = 2
b = 5
f = np.cos(a*x) + b


dx = x[1] - x[0]
df = np.gradient(f, dx)     # first derivative
ddf = np.gradient(df, dx)   # second derivative


fig, axes = plt.subplots(3, 1, figsize=(6,6))

axes[0].plot(x, f)
axes[0].set_title(rf"$f(x) = cos({a}x)+{b}$")
axes[0].set_xlabel("x")
axes[0].set_ylabel(r"$f(x)$")
axes[0].grid()

axes[1].plot(x, df, 'r')
axes[1].set_title("First derivative")
axes[1].set_xlabel("x")
axes[1].set_ylabel(r"$df/dx$")
axes[1].grid()

axes[2].plot(x, ddf, 'g')
axes[2].set_title("Second derivative")
axes[2].set_xlabel("x")
axes[2].set_ylabel(r"$df^{2}/dx^{2}$")
axes[2].grid()

plt.tight_layout()
plt.show()
```
